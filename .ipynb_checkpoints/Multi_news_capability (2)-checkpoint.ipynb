```python
def my_function_body(i, sub1 = None, sub2 = None, word1 = None, word2 = None):
    start_date = i[0]
    end_date = i[1]
    proximity = i[2]
    company_name = i[3]

    body = {
        'load_date_from': start_date,
        'load_date_to': end_date,
        'company': company_name

    }
    if word1 is not None:
        if word2 is not None:
            body['content'] = '"{} {}"~{}'.format(word1, word2, proximity)
        else:
            body['content'] = word1
    if sub1 is not None:
        if sub2 is not None:
            body['subject'] = "\"{}\" AND \"{}\"".format(sub1, sub2)
        else:
            body['subject'] = sub1
    news_renew_content = {}
    try:
        news_thematic = requests.post('http://pa1-ec-lnews001.cloud.msci.net:7001/v1/search/',
                         json=body).json()['articles']
        news_renew_content[company_name] = pd.io.json.json_normalize(news_thematic)
        news_renew_content[company_name]['Company'] = company_name
    except KeyError:
        pass
    time.sleep(1)
    return news_renew_content
```

## function for reading all companies

```python
import requests
API = 'http://pa1-ec-lnews001.cloud.msci.net:7001/v1/search/'
def pull_articles_data(body):
    articles = []
    try:
        articles = requests.post(API, json=body).json()['articles']
    except Exception as ex:
        print("Error occured for body :: {}".format(body))    
    return articles

def pull_data_from_news_repo(info, max_news=None, sub1 = None, sub2 = None, word1 = None, word2 = None):
    start_date = info[0]
    end_date = info[1]
    proximity = info[2]
    news = []
    try:
        body = {
            'load_date_from': start_date,
            'load_date_to': end_date,
        }
        if word1 is not None:
            if word2 is not None:
                body['content'] = '"{} {}"~{}'.format(word1, word2, proximity)
            else:
                body['content'] = word1
        if sub1 is not None:
            if sub2 is not None:
                body['subject'] = "\"{}\" AND \"{}\"".format(sub1, sub2)
            else:
                body['subject'] = sub1

        total_news_found = requests.post(API, json=body).json()['pagination']['totalResults']
        print("Total news found :: {}".format(total_news_found))
        batch_size = 50
        if max_news is not None:
            total_news_found = max_news
        if total_news_found > 0:
            with mp.Pool(processes=12) as pool:
                futures = []
                for i in range(0, int(total_news_found), batch_size):
                    body['offset'] = i
                    body['limit'] = batch_size
                    futures.append(
                        pool.apply_async(pull_articles_data, args=(body,))
                    )
                for f in futures:
                    news.extend(f.get())
    except Exception as ex:
        print("Error occured in request for :: {} :: {}".format(info, ex))
    return news
```

```python
def extractpercent(a, b):
    result = []
    for name in b:
        new_name = re.sub("[\(\[].*?[\)\]]", "", name[:-6]).rstrip()+name[-6:]
        if a in new_name[:-6]:
            result.append(new_name.split('(')[1][:-1])
    if result:
        return result[0]
    else:
        return None
```

## Sentences extraction function

```python
def sentence_extractor(string):
    sentences = nlp(string).sents
    original_sentences = []
    lowercased_sentences = []
    uppercased_sentences = []
    propercased_sentences = []
    lemmatized_sentences = []
    for x in sentences:
        original_sentences.append(x.text)
        lowercased_sentences.append(x.text.lower())
        uppercased_sentences.append(x.text.upper())
        propercased_sentences.append(x.text.title())
        lemmatized_sentences.append(x.lemma_.lower())
    return {'sentences': original_sentences, 'lowercased_sentences': lowercased_sentences,'uppercased_sentences': uppercased_sentences,'propercased_sentences': propercased_sentences, 'lemmatized_sentences': lemmatized_sentences}
def find_sentences_with_keywords(sentence_dict):
    kw_sentence_dict = dict()
    for kw_idx, keyword in enumerate(keywords):
        filtered_sentences = []
        for idx in range (len(sentence_dict['sentences'])):
            words_in_keyword = keyword.split()
            kw_present = True
            for kw in words_in_keyword:
                if not (
                    kw in sentence_dict['sentences'][idx] or 
                    kw in sentence_dict['lowercased_sentences'][idx] or
                    kw in sentence_dict['uppercased_sentences'][idx] or
                    kw in sentence_dict['propercased_sentences'][idx] or
                    kw in sentence_dict['lemmatized_sentences'][idx] or 
                    keywords_lemma[kw_idx] in sentence_dict['sentences'][idx] or 
                    keywords_lemma[kw_idx] in sentence_dict['lowercased_sentences'][idx] or 
                    keywords_lemma[kw_idx] in sentence_dict['uppercased_sentences'][idx] or 
                    keywords_lemma[kw_idx] in sentence_dict['propercased_sentences'][idx] or 
                    keywords_lemma[kw_idx] in sentence_dict['lemmatized_sentences'][idx]
                   ):
                    kw_present = False
            if kw_present:
                filtered_sentences.append(sentence_dict['sentences'][idx])
        kw_sentence_dict[keyword] = filtered_sentences
    return kw_sentence_dict
```

```python
if len(word1)>0:
    sentences_df_list = []
    for n in tqdm_notebook(range(0, len(final_df), 5000)):
        final_df_tmp = final_df.iloc[n:n+5000]
        texts = []
        for i in range(len(final_df_tmp)):
            texts.append((final_df_tmp['id'].iloc[i], final_df_tmp['content'].iloc[i]))
        results = []
        with mp.Pool(processes=mp.cpu_count()) as pool:
            futures = []
            for text_id, text in texts:
                futures.append(
                    (
                        text_id,
                        pool.apply_async(sentence_extractor, args=(text,))
                    )
                )
            results = [(text_id, f.get()) for text_id, f in futures]

        keywords_lemma = []
        for kw in keywords:
            sp_doc = nlp(kw)
            lemma_kw = ''
            for w in sp_doc:
                lemma_kw = lemma_kw + w.lemma_ + ' '
            keywords_lemma.append(lemma_kw.strip())
        kw_sentences = []
        with mp.Pool(processes=mp.cpu_count()) as pool:
            futures = []
            for text_id, sentence_dict in results:
                futures.append(
                    (
                        text_id,
                        pool.apply_async(find_sentences_with_keywords, args=(sentence_dict,))
                    )
                )
            kw_sentences = [[text_id, f.get()] for text_id, f in futures]
        df_list = []
        for i in range(len(kw_sentences)):
            df_word = pd.DataFrame([kw_sentences[i][1]])
            df_word['id'] = kw_sentences[i][0]
            df_list.append(df_word)
        sentence_df = pd.concat(df_list, sort=False, ignore_index = True).fillna(0)
        sentences_df = final_df.reset_index().merge(sentence_df, left_index=True, right_index=True, suffixes=('', '_y'))
        sentences_df = sentences_df.drop(list(sentences_df.filter(regex='_y$')), axis=1)
        sentences_df = sentences_df.applymap(lambda x: "NA" if "[]" in str(x) else x)
        sentences_df_list.append(sentences_df)
    sentences_df = pd.concat(sentences_df_list)
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
```

```python
data = DF.content.values.tolist()

# Remove Emails
data = [re.sub('\S*@\S*\s?', '', sent) for sent in data]

# Remove new line characters
data = [re.sub('\s+', ' ', sent) for sent in data]

# Remove distracting single quotes
data = [re.sub("\'", "", sent) for sent in data]
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(data))
```

```python
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
data_words_nostops = remove_stopwords(data_words)

# Form Bigrams
data_words_trigrams = make_trigrams(data_words_nostops)

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
# python3 -m spacy download en
nlp = spacy.load('en', disable=['parser', 'ner'])

# Do lemmatization keeping only noun, adj, vb, adv
data_lemmatized = lemmatization(data_words_trigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized
new_text = []
for text in texts:
    new_text.append([x for x in text if '_' in x])
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in new_text]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=4, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
  
# Reads 'Youtube04-Eminem.csv' file  
  
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in DF.content: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
